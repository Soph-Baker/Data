```python
NEO_data.plot(kind='scatter', x='Close_Approach_CA_Date', y='CA_Distance_Nominal_km', alpha=0.5, color='black')
plt.xlabel('Close Approach Date')
plt.ylabel('Distance from Earth (km)')
plt.title('NEO Close Approaches Past and Future')
plt.axis('off')
```

```python
NEO_past_data.plot(kind='scatter', x='Close_Approach_CA_Date', y='CA_Distance_Nominal_km', alpha=0.5, color='black')
plt.xlabel('Close Approach Date')
plt.ylabel('Distance from Earth (km)')
plt.title('NEO Close Approaches Past')
```

```python
NEO_future_data.plot(kind='scatter', x='Close_Approach_CA_Date', y='CA_Distance_Nominal_km', alpha=0.5, color='black')
plt.xlabel('Close Approach Date')
plt.ylabel('Distance from Earth (km)')
plt.title('Predicted NEO Close Approaches')
```

```python
NEO_data.CA_Distance_Nominal_km.hist()
plt.title(f"Distance from Centre of Earth km")
plt.xlabel('Distance from earth (km)')
plt.ylabel('Number of NEO')
```

```python
df1 = pd.DataFrame({'Past': NEO_past_data.CA_Distance_Nominal_km})
df2 = pd.DataFrame({'Future': NEO_future_data.CA_Distance_Nominal_km})

df1['Past'].hist(alpha=0.5, label='Past', bins=20, color='Blue')
df2['Future'].hist(alpha=0.5, label='Future', bins=20, color='red')

plt.xlabel('Distance from Centre of Earth (km)')
plt.ylabel('Frequency')
plt.title('Difference Between Frequency of Past and Future NEOs')
plt.legend()
```

```python
NEO_data.plot(kind='scatter', x='V_relative_km/s', y='CA_Distance_Nominal_km', alpha=0.5, color='black')
plt.xlabel('Speed of NEOs (km/s)')
plt.ylabel('Distance from Earth (km)')
plt.title('NEO Close Approaches Speed')
```

```python
NEO_data.plot(kind='scatter', x='V_relative_km/s', y='H_mag', alpha=0.5, color='black')
plt.xlabel('Speed of NEOs (km/s)')
plt.ylabel('Size of NEO')
plt.title('NEO Close Approaches Speed')
```
